```python
# 숫자 데이터를 특정 범위의 값으로 변환 하여 새로운 컬럼에 저장
# 방법 1 : 비교구문과 Loc 함수를 활용하여 파생변수를 생성
# loc [조건, 항목명] = 새로운 값 /
# 해당 조건에 일치하는 데이터들에 대해 '항목명'에 '새로운 값'을 추가
cond1 = df1['Age'] >= 70
cond2 = df1['Age'] >= 60
cond3 = df1['Age'] >= 50
cond4 = df1['Age'] >= 40


df1['Age_Group'] = '30대 이하'
df1.loc[cond4, 'Age_Group'] = '40대'
df1.loc[cond3, 'Age_Group'] = '50대'
df1.loc[cond2, 'Age_Group'] = '60대'
df1.loc[cond1, 'Age_Group'] = '70대 이상'
```

```python
df1['Age_Group']
```

```python
# 파생변수 생성
# 방법 2 : 직접 사용자 함수를 만들어서 특정 데이터를 처리

def func1(row):
    if row >=70:
        return '70대 이상'
    elif row >= 60:
        return '60대'
    elif row >= 50:
        return '50대'
    elif row >= 40:
        return '40대'
    else:
        return '30대 이하'
    # row는 df1['Age']의 각 행의 값이 매개변수
    # row가 70이상이면 '70대 이상'을 반환
    # row가 60이상이면 '60대'를 반환
    # row가 50이상이면 '50대'를 반환
    # row가 40이상이면 '40대'를 반환
    # 그 외에는 '30대 이하'를 반환
    # 각 행의 값에 따라 해당하는 문자열을 반환
    # 이 함수를 df1['Age']의 각 행에 적용하여 새로운 값을 생성
    

df1['Age_Group'] = df1['Age'].apply(func1) # 특정 시리즈의 각 Row 갑을 특정 함수의 매개 변수로 할당

```

```python
# 더 쉽게표현
def func2(row):
    # 연령의 값을 문자로 변환한 뒤, 첫번째 글자만 가져와 연령대로 변환
    return str(row)[0] + '0대'

df1['Age_Group1'] = df1['Age'].apply(func2) # 특정 시리즈의 각 Row 갑을 특정 함수의 매개 변수로 할당
df1['Age_Group1'].value_counts() # 연령대별로 몇명이 있는지 확인
# NaN : 결측치, 데이터가 없는 경우( Not a Number )
```

```python
# 결측값은 그대로 두고싶은 경우
# 함수를 만들 때 결과가 나오도록 만들고 세부적으로 작성

def func3(row):
    # 연령에 포함된 결측갑을 따로 처리
    # pandas에서 결측값을 확인하는 함수가 있다 : pd.isna()
    if pd.isna(row): # 특정값이 결측값이면 True / 아니면 False
        return '미확인'
    elif row >= 100:
        return '90대'
    else:
        # 연령의 값을 문자로 변환한 뒤, 첫번째 글자만 가져와 연령대로 변환
        return str(row)[0] + '0대'
   

df1['Age_Group'] = df1['Age'].apply(func3) 
df1['Age_Group'].value_counts()
```

## 8-2 기술통계량 확인

- 데이터의 특성을 알기 위해 기본적으로 확인하는 통계량
- 연속형 : 
    - 대표값 (평균 / 중앙값) : 해당 연속형 자료를 대표하는 수치
        - 편균의 경우, 이상치(Outlier)에 의해 매우 높거나 낮게 계산될 수 있으므로, 반드시 중위수 (중앙값)와 같이 확인
        - A : 1, 3, 2, 5, 4         평균 : 3
        - A : 1,2,3,4,5             중앙값 : 3

        - B : 1, 3, 2, 4, 100000    평균 : 약 2000
        - B : 1,2,3,4,10000         중앙값 : 3
    
    - 산포 (표준편차) : 해당 연속형 자료의 데이터들이 대표값으로부터 얼마나 떨어져 있는지 나타내는 통계량
        - 산포는 데이터가 얼마나 대표값에 대해 정확한가를 나타내는 척도
        - 데이터가 많이 수집 될 수록, 대표값에 대해 모여있는 분포를 나타냄 (중심극한의 정리)
        - 데이터가 많아지면, 산포가 줄어들 가능성이 있음
    - 분포의 모양 : 해당 연속형 자료의 분포의 모양 / 시각화
        - 정규분포 / 정규분포가 아닌 데이터
        - 분포의 모양에 따라 가설검정 및 통계적 분석 방법이 달라질 수 있음
        - 데이터 수집 관점

- 범주형 : 항목과 빈도수 확인 (Uniqe / Frequency) / 비율

```python
# 월 랜탈빙용 (Amount_Month)과 계약 기간(월) (Term)을 곱하여, 총 계약급을 계산해, 통계량 확인
df1['총계약금'] = df1['Amount_Month'] * df1['Term'] # 월 렌탈 비용과 계약 기간을 곱하여 총 계약금액 계산
df1
```

```python
print(df1['총계약금'].mean()) # 총 계약금의 평균
print(df1['총계약금'].median()) # 총 계약금의 중앙값 
```

```python
# 산포
# 표준편차 : 평균으로부터 얼마나 떨어져 있는지(평균 - 이상치의 영향을 받는 값)
# 범위 : 최대값 - 최소값
# 사분범위 (IQR) : 전체 데이터의 50% 모여있는 구간
print(df1['총계약금'].std()) # 표준편차
print(df1['총계약금'].quantile(0.25)) # 25% 1사분위 값
print(df1['총계약금'].quantile(0.75)) # 75% 3사분위 값

df1['총계약금'].quantile(0.75) - df1['총계약금'].quantile(0.25) # 사분범위 (IQR)
```

```python
df_product.pivot_table(index='회원번호', values=['구매금액', '구매수량'], aggfunc='sum')
```

```python
# 회원번호를 이용해서 데이터를 병합하려고 하면 위와 같이 회원번호가 Index로 되어있으면 안됨.
df_product.pivot_table(index='회원번호', values=['구매금액', '구매수량'], aggfunc='sum').reset_index()
# .reset_index() : Index 값을 0 부터 다시 부여
```

```python
p1 = df_product.pivot_table(index='회원번호',
                            values=['구매금액', '구매수량'], aggfunc='sum').reset_index()
p1.shape
```

## 데이터 테이블 병합(Table Join)


```python
# Key Column(회원번호)에 공통된 값만 가져와 병합
m1 = pd.merge(df_member, p1, on='회원번호', how='inner')
# how = 테이이블을 어떤 테이블 기준으로 병합할 것인지  / inner /outer / Left / Right
# on = Key Column
print(m1.shape)
m1.head(2)
```

```python
# 한명의 회원이 몇번 방문 했는가 계산
df_product['Label'] = 1 # Label에 '1'이라고 붙임.
```

```python
p2 = df_product.pivot_table(index=['회원번호','공급일자'], 
                       values='Label', aggfunc='min').reset_index()
# 최소값으로 보여지게 'min' 
# 회원번호와 공급일자의 데이터를 사용하기 위해서 reset_index() 사용
# 공급일자에 회원이 왔다면 Label에 1 이라는 값이 출력됨
# 정리 //
# 한명의 고객 별로 언제 방문했는가를 1값으로 식별힉 위해, 회원-날짜-1 계산
p2.head()
```

```python
p2.pivot_table(index='회원번호', values='Label', aggfunc='sum')
# 회원이 해당 기간동안 몇번 방문했는지 확인
```

```python
p3 = p2.pivot_table(index='회원번호', values='Label', aggfunc='sum').reset_index()
print(p3.shape)
p3.head()
```

```python
# m1과 p3 병합
pd.merge(m1, p3, on='회원번호', how='inner')
```

```python
p4 = pd.merge(m1, p3, on='회원번호', how='inner')
p4.rename(columns={'구매금액':'총구매금액', '구매수량':'총구매수량', 'Label':'총방문횟수'})
```

```python
df5 = p4.rename(columns={'구매금액':'총구매금액', '구매수량':'총구매수량', 'Label':'총방문횟수'})
print(df5.shape)
df5.head(3)
```
